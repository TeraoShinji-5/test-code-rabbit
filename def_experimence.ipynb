以下のコードのエンコード機能とデコード機能を関数化して、<br>
"componemts"ディレクトリ内の"encode_to_decode.py"に保存してみよう。

```python
encoded_data = ""
for category in category_array:
    encoded_data += "1" if categories.get(category, False) else "0"

print("エンコードデータ", encoded_data)   

decoded_data = {}
query_array = []
for i in range(len(category_array)):
    if encoded_data[i] == "1":
        decoded_data[category_array[i]] = True
        query_array.append(category_array[i])
    else:
        decoded_data[category_array[i]] = False

print("デコードデータ", decoded_data)

true_categories = [key for key, value in decoded_data.items() if value]

print("選択したのカテゴリ一覧:", true_categories)
```
